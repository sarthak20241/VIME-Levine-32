```python
results = pd.DataFrame(
        columns=['p_m', 'alpha', 'K', 'beta', 'label_no', 'Accuracy', 'F1 Score', 'Precision', 'Recall Score', 'AUROC'])

hyperparameters = {
        'p_m': [0.2, 0.3],
        'alpha': [1.0, 2.0],
        'K': [3],
        'beta': [0.6, 0.8],
        'label_no': [250]
    }

    # iterate over all possible combinations of hyperparameters
for p_m in hyperparameters['p_m']:
        for alpha in hyperparameters['alpha']:
            for K in hyperparameters['K']:
                for beta in hyperparameters['beta']:
                    for label_no in hyperparameters['label_no']:
                        # run the model with the current hyperparameters
                        result = run(p_m, alpha, K, beta, label_no)
                        results = results.append({'p_m': p_m,
                                                  'alpha': alpha,
                                                  'K': K,
                                                  'beta': beta,
                                                  'label_no': label_no,
                                                  'Accuracy': result[0][4],
                                                  'F1 Score': result[1][4],
                                                  'Precision': result[2][4],
                                                  'Recall Score': result[3][4],
                                                  'AUROC': result[4][4]}, ignore_index=True)
```
